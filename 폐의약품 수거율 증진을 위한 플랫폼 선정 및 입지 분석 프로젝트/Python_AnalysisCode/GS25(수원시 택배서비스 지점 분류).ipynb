## 데이터 전처리

- '수원'& '택배서비스' 지점 선별
- Tableau 맵 시각화를 위한 address columns 분할
- 자치구별 GS25 편의점 개수 파악

```python
#결측치 확인
df.isnull().sum()
```

```python
# '수원'시를 제외한 다른 지역 편의점 데이터 삭제
df = df[df['address'].str.startswith('경기 수원')]
```

#### Tableau 필터 기능을 위한 '시','구' column 생성
- address_gu 타지역 데이터는 무시함

```python
# 지역 column 생성
def add_column1(df):
    df['address_si'] = df['address'].apply(lambda x: x.split()[1])
    
    return df
```

```python
df = add_column1(df)
```

```python
# 서울시 '구' column 생성, 타지역은 무관
def add_column2(df):
    df['address_gu'] = df['address'].apply(lambda x: x.split()[2])
    
    return df
```

```python
df = add_column2(df)
```

```python
df.head()
```

```python
#address_si 이동
def move_address_si(df):
    address_si = df['address_si']
    df = df.drop(columns=['address_si'])
    df.insert(3, 'address_si', address_si)
    return df

df = move_address_si(df)
```

```python
#address_gu 이동
def move_address_gu(df):
    address_gu = df['address_gu']
    df = df.drop(columns=['address_gu'])
    df.insert(4, 'address_gu', address_gu)
    return df

df = move_address_gu(df)
```

```python
#longs columns 이동
def move_longs(df):
    longs = df['longs']
    df = df.drop(columns=['longs'])
    df['longs'] = longs
    return df

df = move_longs(df)
```

```python
#lat,longs columns 명 변경
df = df.rename(columns={'lat': 'latitude', 'longs': 'longitude'})
```

```python
#인덱스 재설정
df.reset_index(drop=True, inplace=True)
```

```python
# address_gu column 카운팅
def add_column3(df):
    gu_counts = df['address_gu'].value_counts()
    df['count_address_gu'] = df['address_gu'].map(gu_counts)
    return df
```

```python
df = add_column3(df)
```

```python
#offeringService, 서비스를 제공하지 않는 (NaN) 데이터 삭제
df = df.dropna(subset=['offeringService'])
```

```python
#택배 서비스를 수행하는 GS25 편의점 지점 개수 확인
count = len(df[df['offeringService'].apply(lambda x: 'post' in x)])
print(count)
```

```python
#택배 서비스가 없는 GS25 편의점 지점 삭제
df = df[df['offeringService'].apply(lambda x: 'post' in x)]
```

```python
df.shape
```

```python
#논리연산자를 사용한 'post' 데이터 유무 확인
def check_post(df):
    post_gs25 = df[~df['offeringService'].str.contains('post', na=False)]
    return post_gs25

post_gs25 = check_post(df)
print(post_gs25)
```
